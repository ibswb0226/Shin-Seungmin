```python
cause_of_deaths = [col for col in df.columns if col not in ('Country/Territory', 'Code', 'Year')] #불필요한 열 제외

df['Total_deaths'] = df[cause_of_deaths].sum(axis=1) 
```

```python
disease_df = df[cause_of_deaths].sum().to_frame().reset_index() #질병명과 그에 대한 전체수 확인
disease_df.rename(columns={"index": "Disease", 0:"Total cases"}, inplace=True)
disease_df
```

```python
country_group_df=df.drop(['Year','Code'],axis=1) #사용하지 않는 컬럼 제거
country_group_df.groupby('Country/Territory').sum() #국가별로 질병 수 확인
```

```python
# 중간 보고서에서는 가장 높은 사망원인 그래프의 시각화가 잘못되어 최종단계에서 그래프를 수정하였습니다.

#사망수가 가장 높은 질병 퍼센티지


fig = px.sunburst(disease_df, 
                  path=['Disease'], 
                  values='Total cases'
                  )

fig.update_traces(textinfo='label+percent parent')    
fig.update_layout(title_text='Percentage of Cause of Deaths', title_x=0.5, font_size=15)
fig.show()
```
