```python
cause_of_deaths = [col for col in df.columns if col not in ('Country/Territory', 'Code', 'Year')]

df['Total_deaths'] = df[cause_of_deaths].sum(axis=1)
```

```python
disease_df = df[cause_of_deaths].sum().to_frame().reset_index()
disease_df.rename(columns={"index": "Disease", 0:"Total cases"}, inplace=True)
disease_df
```

```python
pd.DataFrame(df[cause_of_deaths].sum())[0:10].sort_values(0,ascending=False)\
.plot(kind='barh',xlabel='Diseases',ylabel='Deaths',title='TOP 10 Diseases By Deaths(1990-2019)',legend=False,figsize=(15,5))
```

```python
Top10_deaths = df.groupby('Country/Territory')['Total_deaths'].sum().sort_values(ascending=False).head(10).reset_index()

Top10_deaths
```

```python
plt.figure(figsize=(16,9))

sns.barplot(data = Top10_deaths, 
            x = 'Country/Territory', 
            y = 'Total_deaths', 
            palette = 'pastel'
            )

plt.xticks(rotation = 90)
plt.xlabel('Country', fontsize = 12)
plt.ylabel('Total Number of Deaths', fontsize = 12)
plt.title('Top 10 Countries with the Highest Number of Deaths', fontsize =15)
```

```python
Low10_deaths = df.groupby('Country/Territory')['Total_deaths'].sum().sort_values(ascending=True).head(10).reset_index()

Low10_deaths
```

```python
plt.figure(figsize=(16,9))

sns.barplot(data = Low10_deaths, 
            x = 'Country/Territory', 
            y = 'Total_deaths', 
            palette = 'pastel'
            )

plt.xticks(rotation = 90)
plt.xlabel('Country/Territory', fontsize = 12)
plt.ylabel('Total Number of Deaths', fontsize = 12)
plt.title('Top 10 Countries with the Lowest Number of Deaths', fontsize =15)
```

```python
def top_10_illnesses(country):
    df_country = df[df['Country/Territory'] == country]
    min_year, max_year = df_country['Year'].min(), df_country['Year'].max()
    #print(df_country.head())
    fig = px.bar(df_country.iloc[:, 3:].sum().sort_values(ascending=False).head(10)[3:], title='Top 10 Cause of Deaths in {} between {} and {}'.format(country, min_year, max_year), color_discrete_sequence=px.colors.qualitative.Pastel, labels={'value': 'Amount of deaths', 'index': 'Cause of Deaths'})
    fig.update_layout(showlegend=False)
    fig.show()
```
